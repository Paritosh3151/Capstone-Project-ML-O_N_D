Let's have a closer look at the distribution of temperature and ph.
It is symmetrical and bell shaped, showing that trials will usually give a result near the average, but will occasionally
deviate by large amounts. It's also fascinating how these two really resemble each other!

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
# sns.distplot(df_setosa['sepal_length'],kde=True,color='green',bins=20,hist_kws={'alpha':0.3})
sns.distplot(c_p_VF_df['Temperature'],color="purple",bins=15,hist_kws={'alpha':0.2})
plt.subplot(1, 2, 2)
sns.distplot(c_p_VF_df['Rainfall'],color="green",bins=15,hist_kws={'alpha':0.2})
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
# sns.distplot(df_setosa['sepal_length'],kde=True,color='green',bins=20,hist_kws={'alpha':0.3})
sns.distplot(c_p_VF_df['Temperature'],color="purple",bins=15,hist_kws={'alpha':0.2})
plt.subplot(1, 2, 2)
sns.distplot(c_p_VF_df['Ph'],color="green",bins=15,hist_kws={'alpha':0.2})
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
# sns.distplot(df_setosa['sepal_length'],kde=True,color='green',bins=20,hist_kws={'alpha':0.3})
sns.distplot(c_p_VF_df['Temperature'],color="purple",bins=15,hist_kws={'alpha':0.2})
plt.subplot(1, 2, 2)
sns.distplot(c_p_VF_df['Production'],color="green",bins=15,hist_kws={'alpha':0.2})
```

```python
c_p_VF_df.groupby(['Cropconversion','Crop'])['Temperature'].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Cropconversion')
plt.show()
```

```python
c_p_VF_df.groupby(['Cropconversion','Crop'])['Rainfall'].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Cropconversion')
plt.show()
```

```python
c_p_VF_df.groupby(['Cropconversion','Crop'])['Ph'].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Cropconversion')
plt.show()
```

```python
c_p_VF_df.groupby(['Cropconversion','Crop'])['Production'].sum().sort_values(ascending = True).plot(kind = 'barh',figsize = (13,13))
plt.tight_layout()
plt.title('Cropconversion')
plt.show()
```

```python
# pairplot based on state
sns.pairplot(c_p_VF_df[['Cropconversion','Crop','Rainfall','Temperature','Ph','Production']], hue = 'Cropconversion')
plt.tight_layout()
plt.show()
```
